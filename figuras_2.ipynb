```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el círculo original
radio = 1
theta = np.linspace(0, 2*np.pi, 100)
x_circulo = radio * np.cos(theta)
y_circulo = radio * np.sin(theta)

# Definir el vector de traslación
tx = 2  # traslación en el eje x
ty = 1  # traslación en el eje y

# Aplicar la transformación de traslación al círculo
x_traslado = x_circulo + tx
y_traslado = y_circulo + ty

# Graficar el círculo original y su traslación en la misma figura
plt.figure(figsize=(6, 6))
plt.plot(x_circulo, y_circulo, 'b-', label='Círculo original')
plt.fill(x_circulo, y_circulo, color='lightblue', alpha=0.5)
plt.plot(x_traslado, y_traslado, 'r-', label='Círculo trasladado')
plt.fill(x_traslado, y_traslado, color='lightcoral', alpha=0.5)
plt.title('Círculo y su traslación')
plt.grid(True)
plt.axis('equal')
plt.legend()

plt.show()
```
