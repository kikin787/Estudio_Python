```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def func(x):
    return 2 * (x**2) * np.cos(x) - 5 * x

def func_derivative(x):
    return 4 * x * np.cos(x) - 2 * (x**2) * np.sin(x) - 5

# Parámetros del algoritmo
learning_rate = 0.01
num_iterations = 1000
initial_x = 2.0  # Valor inicial

# Inicializar
x = initial_x
x_vals = [x]  # Lista para almacenar los valores de x en cada iteración

# Gradiente Descendente
for i in range(num_iterations):
    gradient = func_derivative(x)
    x = x - learning_rate * gradient
    x_vals.append(x)

# Crear valores de x para la gráfica
x_range = np.linspace(-2, 4, 400)
y_range = func(x_range)

# Visualizar la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='Función $2x^2 \cos(x) - 5x$')
plt.scatter(x_vals, [func(x) for x in x_vals], color='red', label='Puntos de iteración')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Método del Gradiente Descendente')
plt.legend()
plt.grid()
plt.show()

# Imprimir el resultado
print("Valor mínimo aproximado de la función:", func(x))
print("Valor de x donde se encuentra el mínimo:", x)
```
