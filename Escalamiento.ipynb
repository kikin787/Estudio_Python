```python
import cv2
import matplotlib.pyplot as kikin

# Cargar la imagen
imagen = cv2.imread('home.jpg')

# Función para mostrar la imagen original y la escalada juntas, con un punto específico
def mostrar_escalamiento_con_punto(titulo, imagen_original, imagen_escala, punto=None, nuevo_punto=None):
    fig, axs = plt.subplots(1, 2, figsize=(12, 6))
    axs[0].imshow(cv2.cvtColor(imagen_original, cv2.COLOR_BGR2RGB))
    axs[0].set_title('Original')
    axs[1].imshow(cv2.cvtColor(imagen_escala, cv2.COLOR_BGR2RGB))
    axs[1].set_title(titulo)
    
    # Mostrar el punto específico si está disponible
    if punto is not None and nuevo_punto is not None:
        axs[1].plot(punto[0], punto[1], 'ro', markersize=5)
        axs[1].plot(nuevo_punto[0], nuevo_punto[1], 'go', markersize=5)
        axs[1].annotate('Punto Original', punto, xytext=(punto[0]+10, punto[1]-10), color='r')
        axs[1].annotate('Nuevo Punto', nuevo_punto, xytext=(nuevo_punto[0]+10, nuevo_punto[1]-10), color='g')

    kikin.show()


# Función para el escalamiento uniforme
def escalamiento_uniforme(imagen, escala):
    alto, ancho = imagen.shape[:2]
    nuevo_alto = int(alto * escala)
    nuevo_ancho = int(ancho * escala)
    imagen_esc_uniforme = cv2.resize(imagen, (nuevo_ancho, nuevo_alto))
    return imagen_esc_uniforme


# Función para el escalamiento diferencial
def escalamiento_diferencial(imagen, escala_x, escala_y):
    alto, ancho = imagen.shape[:2]
    nuevo_alto = int(alto * escala_y)
    nuevo_ancho = int(ancho * escala_x)
    imagen_esc_diferencial = cv2.resize(imagen, (nuevo_ancho, nuevo_alto))
    return imagen_esc_diferencial


# Función para el escalamiento a un punto específico
def escalamiento_punto(imagen, escala, punto):
    alto, ancho = imagen.shape[:2]
    nuevo_ancho = int(ancho * escala)
    nuevo_alto = int(alto * escala)
    imagen_esc_punto = cv2.resize(imagen, (nuevo_ancho, nuevo_alto), interpolation=cv2.INTER_CUBIC)

    # Corregir la posición del punto
    nuevo_punto = (int(punto[0] * escala), int(punto[1] * escala))

    return imagen_esc_punto, nuevo_punto


# Escalar uniformemente
imagen_esc_uniforme = escalamiento_uniforme(imagen, 1.5)
mostrar_escalamiento_con_punto('Escalamiento Uniforme', imagen, imagen_esc_uniforme)

# Escalar diferencialmente
imagen_esc_diferencial = escalamiento_diferencial(imagen, 2, 0.5)
mostrar_escalamiento_con_punto('Escalamiento Diferencial', imagen, imagen_esc_diferencial)

# Escalar a un punto específico
punto_especifico = (200, 300)
imagen_esc_punto, nuevo_punto = escalamiento_punto(imagen, 0.8, punto_especifico)
mostrar_escalamiento_con_punto('Escalamiento a Punto Específico', imagen, imagen_esc_punto, punto_especifico, nuevo_punto)
```
