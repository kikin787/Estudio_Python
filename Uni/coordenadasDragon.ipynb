```python
import matplotlib.pyplot as plt
import numpy as np

#función para cargar las coordenadas desde el archivo
def load_coordinates_from_file(file_path):
    coordinates = []
    with open(file_path, 'r') as file:
        for line in file:
            x, y, z = map(int, line.strip().split())
            coordinates.append((x, y, z))
    return coordinates

#Ruta al archivo de coordenadas
file_path = "MatrizCapas_limpio.txt"

#Cargar las coordenadas desde el archivo
coordinates = load_coordinates_from_file(file_path)

#Tomar un subconjunto aleatorio de las coordenadas
sample_size = 100000 #Puedes ajustar este valor según tus necesidades
random_indices = np.random.choice(len(coordinates), sample_size, replace=False)
sampled_coordinates = [coordinates[i] for i in random_indices] 

#Separar las coordenadas en listas separadas para x, y, z
x_coords = [coord[0] for coord in sampled_coordinates]
y_coords = [coord[1] for coord in sampled_coordinates]
z_coords = [coord[2] for coord in sampled_coordinates]

#Crear figura y ejes 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

#Graficar las coordenadas en 3D
ax.scatter(x_coords, y_coords, z_coords, c=z_coords,cmap='viridis', marker='o')

#Etiquetas de los ejes
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.title('Sampled Coordinates (3D)')

plt.show()
```
