```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la funci칩n y su derivada
def func(x):
    return 2 * x * x * np.cos(x) - 5 * x
def func_derivative(x):
    return 4 * x * np.cos(x) - 2 * (x**2) * np.sin(x) - 5
```

```python

# Par치metros del algoritmo
learning_rate = 0.05
num_iterations = 20
initial_x = -1  # Valor inicial

# Inicializar
x = initial_x
x_vals = [x]  # Lista para almacenar los valores de x en cada iteraci칩n

# Gradiente Descendente
for i in range(num_iterations):
    gradient = func_derivative(x)
    x = x - learning_rate * gradient
    x_vals.append(x)

# Crear valores de x para la gr치fica
x_range = np.linspace(-5, 5, 400)
y_range = func(x_range)
```

```python
# Visualizar la gr치fica
plt.figure(figsize=(10, 6))
plt.plot(x_range, y_range, label='Funci칩n $2x^2 \cos(x) - 5x$')
plt.scatter(x_vals, [func(x) for x in x_vals], color='red', label='Puntos de iteraci칩n')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('M칠todo del Gradiente Descendente')
plt.legend()
plt.grid()
plt.show()

# Imprimir el resultado
print("Valor m칤nimo aproximado de la funci칩n:", func(x))
print("Valor de x donde se encuentra el m칤nimo:", x)
```
