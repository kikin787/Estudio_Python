```python
import cv2
import numpy as np
import matplotlib.pyplot as kikin
from skimage.morphology import skeletonize

# Leer la imagen en escala de grises
image = cv2.imread('GOAT.jpg', cv2.IMREAD_GRAYSCALE)

# Invertir la imagen si el fondo es blanco
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY_INV)

# Convertir la imagen binaria a valores booleanos (necesario para skeletonize)
binary_image = binary_image > 0

# Aplicar el adelgazamiento (skeletonization)
skeleton = skeletonize(binary_image)

# Convertir la imagen esqueleto a uint8 para mostrarla
skeleton = (skeleton * 255).astype(np.uint8)

# Graficar la imagen original y el resultado del adelgazamiento
kikin.figure(figsize=(10, 5))

kikin.subplot(1, 2, 1)
kikin.title('Imagen Original')
kikin.imshow(image, cmap='gray')
kikin.axis('off')

kikin.subplot(1, 2, 2)
kikin.title('Adelgazamiento')
kikin.imshow(skeleton, cmap='gray')
kikin.axis('off')

kikin.tight_layout()
kikin.show()
```
