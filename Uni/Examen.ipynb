```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Función de costo a optimizar: f(x, y) = x^2 + 3y
def cost_function(x, y):
    return x**2 + 3*y
```

```python
# Algoritmo Genético
def genetic_algorithm():
    # Implementa tu algoritmo genético aquí
    # Devuelve las mejores soluciones para x e y
    
    best_x = np.random.rand() * 10  # Genera solución aleatoria para x
    best_y = np.random.rand() * 10  # Genera solución aleatoria para y
    return best_x, best_y
```

```python
# Gradiente Descendente
def gradient_descent(initial_x, initial_y, learning_rate, num_iterations):
    current_x = initial_x
    current_y = initial_y
    
    x_vals = []  # Lista para almacenar los valores de x en cada iteración
    y_vals = []  # Lista para almacenar los valores de y en cada iteración
    
    for i in range(num_iterations):
        gradient_x = 2 * current_x  # Derivada parcial de f(x, y) con respecto a x
        gradient_y = 3  # Derivada parcial de f(x, y) con respecto a y
        
        current_x -= learning_rate * gradient_x
        current_y -= learning_rate * gradient_y
        
        # Verificar si los valores de x e y están dentro del rango de la cuadrícula
        if -10 <= current_x <= 10 and -10 <= current_y <= 10:
            x_vals.append(current_x)
            y_vals.append(current_y)
        
    return x_vals, y_vals
```

```python
# Crear una cuadrícula de valores para x e y
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Calcular los valores de f(x, y) para cada punto de la cuadrícula
Z = cost_function(X, Y)

# Ejecuta el algoritmo genético para obtener soluciones iniciales para x e y
initial_x, initial_y = genetic_algorithm()

# Ejecuta el gradiente descendente desde las soluciones iniciales obtenidas
learning_rate = 0.1
num_iterations = 100
x_vals, y_vals = gradient_descent(initial_x, initial_y, learning_rate, num_iterations)

# Graficar la superficie de la función de costo en 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie
surf = ax.plot_surface(X, Y, Z, cmap='viridis')

# Agregar puntos para las soluciones iniciales y el recorrido Gradiente Descendente
ax.scatter(initial_x, initial_y, cost_function(initial_x, initial_y), c='r', marker='o', label='Solución Inicial')
ax.scatter(x_vals, y_vals, [cost_function(x, y) for x, y in zip(x_vals, y_vals)], c='g', marker='s', label='Recorrido Gradiente Descendente', alpha=0.5)

# Configuración de etiquetas
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f(x, y)')

# Título
plt.title('Gráfico en 3D de f(x, y) = x^2 + 3y con Soluciones Iniciales y Recorrido Gradiente Descendente')

# Leyenda
ax.legend()

# Mostrar el gráfico
plt.show()
```
