## Librerías

```python
import matplotlib.pyplot as kikin
import numpy as np
```

## Cargar la imagen

```python
# Cargar la imagen que contiene las figuras geométricas
imagen = kikin.imread('img/horse.jpg')
```

## Parametros de traslación

```python
# Definir la traslación vertical y horizontal
ty = 100  # traslación en el eje y
tx = -30  # traslación en el eje x
```

## Creación de nueva imagen

```python

# Crear una nueva imagen con las mismas dimensiones que la original
nueva_imagen = np.zeros_like(imagen)

```

## Asignación de la región en donde estará la nueva imagen

```python
# Determinar la región donde colocar la imagen trasladada
inicio_y = max(0, ty)  # para evitar valores negativos
fin_y = min(nueva_imagen.shape[0], imagen.shape[0] + ty)
inicio_x = max(0, tx)
fin_x = min(nueva_imagen.shape[1], imagen.shape[1] + tx)
```

## Asignación para colocar nueva imagen poniendo un fondo negro

```python

# Colocar la imagen trasladada en la nueva imagen
nueva_imagen[inicio_y:fin_y, inicio_x:fin_x] = imagen[max(-ty, 0):min(imagen.shape[0], nueva_imagen.shape[0]-ty),
                                                       max(-tx, 0):min(imagen.shape[1], nueva_imagen.shape[1]-tx)]
```

## Impresión de las imagenes

```python
# Mostrar la imagen original y la imagen con la traslación
kikin.figure(figsize=(10, 5))

kikin.subplot(1, 2, 1)
kikin.imshow(imagen, cmap='gray')
kikin.title('Imagen original')

kikin.subplot(1, 2, 2)
kikin.imshow(nueva_imagen, cmap='gray')
kikin.title('Imagen con traslación')

kikin.show()
```
