```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definir la función y sus derivadas parciales
def func(x, y):
    return 2 * x * x * np.cos(x) - 5 * x

def func_partial_x(x, y):
    return 4 * x * np.cos(x) - 2 * x * x * np.sin(x) - 5

def func_partial_y(x, y):
    return 0  # No hay componente en y para esta función

# Parámetros del algoritmo
learning_rate = 0.1
num_iterations = 20
initial_x = 2  # Valor inicial para x
initial_y = 3  # Valor inicial para y

# Inicializar
x = initial_x
y = initial_y
x_vals = [x]
y_vals = [y]

# Gradiente Descendente
for i in range(num_iterations):
    gradient_x = func_partial_x(x, y)
    gradient_y = func_partial_y(x, y)
    x = x - learning_rate * gradient_x
    y = y - learning_rate * gradient_y
    x_vals.append(x)
    y_vals.append(y)

# Crear malla de valores para la gráfica
x_range = np.linspace(-5, 5, 100)
y_range = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = func(X, Y)

# Visualizar la gráfica en 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)
ax.scatter(x_vals, y_vals, [func(x, y) for x, y in zip(x_vals, y_vals)], color='red', label='Puntos de iteración')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Método del Gradiente Descendente en 2D')
plt.legend()
plt.show()

# Imprimir el resultado
print("Valor mínimo aproximado de la función:", func(x, y))
print("Valores de x, y donde se encuentra el mínimo:", x, y)
```
