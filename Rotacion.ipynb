```python
import numpy as np
import matplotlib.pyplot as plt

def rotate_2d(punto, angulo, center):
    """Rotates a 2D point around a center."""
    x, y = punto
    cx, cy = center
    angle_rad = np.radians(angulo)
    x_rotated = cx + np.cos(angle_rad) * (x - cx) - np.sin(angle_rad) * (y - cy)
    y_rotated = cy + np.sin(angle_rad) * (x - cx) + np.cos(angle_rad) * (y - cy)
    return x_rotated, y_rotated

def plot_figure_2d(points, rotated_points, center):
    """Plots 2D figure and its rotation."""
    plt.figure()
    plt.plot([p[0] for p in points], [p[1] for p in points], label='Original')
    plt.plot([p[0] for p in rotated_points], [p[1] for p in rotated_points], label='Rotated')
    plt.plot(center[0], center[1], 'ro', label='Rotation Center', color="purple")  # Plot rotation center in yellow
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('2D Rotation')
    plt.legend()
    plt.axis('equal')
    plt.show()


# Define points for a 2D figure
points_2d_figure = [(-1, 1), (-1, -4), (-5, -4), (-5, 2), (0, 5), (5, 2), (5, -4), (1, -4), (1, 1), (-1, 1)]

# Define rotation center for 2D
center_2d = (0, 0)

# Example input
angle_2d = 90

# Rotate 2D figure
rotated_points_2d_figure = [rotate_2d(point, angle_2d, center_2d) for point in points_2d_figure]
plot_figure_2d(points_2d_figure, rotated_points_2d_figure, center_2d)
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

def rotate_3d(point, angle, center):
    """Rotates a 3D point around a center."""
    x, y, z = point
    cx, cy, cz = center
    angle_rad = np.radians(angle)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad), 0],
                                [np.sin(angle_rad), np.cos(angle_rad), 0],
                                [0, 0, 1]])
    rotated_point = np.dot(rotation_matrix, np.array([x - cx, y - cy, z - cz])) + np.array([cx, cy, cz])
    return rotated_point

def plot_figure_3d(points, rotated_points, center):
    """Plots 3D figure and its rotation."""
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    
    # Plot original points
    ax.plot([p[0] for p in points], [p[1] for p in points], [p[2] for p in points], label='Original', color='blue')
    
    # Plot rotated points
    ax.plot([p[0] for p in rotated_points], [p[1] for p in rotated_points], [p[2] for p in rotated_points], label='Rotated', color='red')
    
    # Plot rotation center
    ax.scatter(center[0], center[1], center[2], label='Rotation Center', color='yellow')
    
    # Connect points in parallel at the same level
    for i in range(0, len(points)//2):
        if points[i] != (-1, 4, 0) and points[i+len(points)//2] != (-1, 2, 0.5):
            ax.plot([points[i][0], points[i+len(points)//2][0]], 
                    [points[i][1], points[i+len(points)//2][1]], 
                    [points[i][2], points[i+len(points)//2][2]], color='blue')
            ax.plot([rotated_points[i][0], rotated_points[i+len(points)//2][0]], 
                    [rotated_points[i][1], rotated_points[i+len(points)//2][1]], 
                    [rotated_points[i][2], rotated_points[i+len(points)//2][2]], color='red')

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')
    ax.set_title('3D Rotation')
    ax.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), shadow=True, ncol=2)
    plt.show()

# Define points for a 3D figure
points_3d_figure = [
    (-1, 1, 0), (-1, -4, 0), (-5, -4, 0), (-5, 2, 0), (0, 5, 0), (5, 2, 0), (5, -4, 0), (1, -4, 0), (1, 1, 0),
    (-1, 1, 0.5), (-1, -4, 0.5), (-5, -4, 0.5), (-5, 2, 0.5), (0, 5, 0.5), (5, 2, 0.5), (5, -4, 0.5), (1, -4, 0.5), (1, 1, 0.5)
]

# Define rotation center for 3D
center_3d = (0, 0, 0)

# Example input
angle_3d = 90

# Rotate 3D figure
rotated_points_3d_figure = [rotate_3d(point, angle_3d, center_3d) for point in points_3d_figure]
plot_figure_3d(points_3d_figure, rotated_points_3d_figure, center_3d)
```
